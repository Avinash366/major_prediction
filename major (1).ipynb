```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_excel(r"C:\Users\ponug\Downloads\Changed Cardio train.xlsx")
df
```

```python
df.isnull().sum()
```

```python
df1=df.drop("height",axis=1)
df1
```

```python
x=df1.iloc[:,1:11]
x
```

```python
y=df1.iloc[:,11:]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.8)
```

```python
models=[LogisticRegression(),RandomForestClassifier(),DecisionTreeClassifier(),KNeighborsClassifier(),SVC()]
for i in range(5):
    models[i].fit(xtrain,ytrain)
    print(f'{models[i]} : ')
    train_pred=models[i].predict(xtest)
    print("score is:",models[i].score(xtest,ytest))
```

```python
classifiers=[LogisticRegression(),RandomForestClassifier(),DecisionTreeClassifier(),KNeighborsClassifier(),SVC()]
best_classifier=None
best_classifier_predictions=None
best_accuracy=None
for classifier in classifiers:
    classifier.fit(xtrain,ytrain)
    predictions=classifier.predict(xtest)
    accuracy=classifier.score(xtest,ytest)
    if best_classifier is None or accuracy>best_accuracy:
        best_classifier=classifier
        best_accuracy=accuracy
if isinstance(best_classifier,RandomForestClassifier):
    rf_classifier=RandomForestClassifier()
    rf_classifier.fit(xtrain,ytrain)
    best_classifier_predictions=rf_classifier.predict(xtest)
elif isinstance(best_classifier,LogisticRegression):
    lr_classifier=LogisticRegression()
    lr_classifier.fit(xtrain,ytrain)
    best_classifier_predictions=lr_classifier.predict(xtest)
elif isinstance(best_classifier,KNeighborsClassifier):
    knn_classifier=KNeighborsClassifier()
    knn_classifier.fit(xtrain,ytrain)
    best_classifier_predictions=knn_classifier.predict(xtest)
elif isinstance(best_classifier,SVC):
    svc_classifier=SVC()
    svc_classifier.fit(xtrain,ytrain)
    best_classifier_predictions=svc_classifier.predict(xtest)
elif isinstance(best_classifier,DecisionTreeClassifier):
    dtc_classifier=DecisionTreeClassifier()
    dtc_classifier.fit(xtrain,ytrain)
    best_classifier_predictions=dtc_classifier.predict(xtest)
print("Prediction based on the best classifier:",best_classifier_predictions,best_classifier,best_accuracy)
```
